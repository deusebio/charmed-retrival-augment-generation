```python
import os
from openai import OpenAI

# 1. Initialize the client pointing to your KServe ingress URL
# (Replace with your actual cluster IP or domain name)
client = OpenAI(
    base_url="http://10.152.183.95/openai/v1",
    api_key="fake-key-kserve-doesnt-require-one"  # A key is required by the SDK but ignored by KServe
)

try:
    # 2. Submit a chat completion request
    response = client.chat.completions.create(
        # model="Qwen/Qwen2.5-3B-Instruct",  # Must match the exact storageUri repo name
        model="qwen25-3b-native",
        messages=[
            {"role": "system", "content": "You are a 17th century poet."},
            {"role": "user", "content": "Create a poem about Ubuntu and its community."},
        ],
        temperature=0.7,
        max_tokens=200
    )

    # 3. Print the text response
    print("Response:")
    print(response.choices[0].message.content)

except Exception as e:
    print(f"An error occurred: {e}")
```
